```python
import pandas as pd
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch
import seaborn as sns
```

```python
import pandas as pd

# Define the file path in Google Drive
file_path = '/content/drive/My Drive/messibetis.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)
```

```python
#convert the data to match the mplsoccer statsbomb pitch
df['x'] = df['x']*1.2
df['y'] = df['y']*.8
df['endX'] = df['endX']*1.2
df['endY'] = df['endY']*.8
```

```python
df
```

```python
# Create the figure and axis
fig, ax = plt.subplots(figsize=(13.5, 8))
fig.set_facecolor('#22312b')
ax.patch.set_facecolor('#22312b')

# Create the pitch with the desired configuration
pitch = Pitch(pitch_type='statsbomb', pitch_color='#22312b', line_color='#c7d5cc')

# Draw the pitch on the ax figure
pitch.draw(ax=ax)

# Invert the x-axis for horizontal orientation
plt.gca().invert_xaxis()

# Create the heatmap using seaborn
sns.kdeplot(
    x=df['x'],
    y=df['y'],
    fill=True,
    shade_lowest=False,
    alpha=0.5,
    cmap='magma',
    ax=ax
)

# Use a for loop to plot each pass
for index, row in df.iterrows():
    if row['outcome'] == 'Successful':
        plt.plot((row['x'], row['endX']), (row['y'], row['endY']), color='green')
        plt.scatter(row['x'], row['y'], color='green')
    elif row['outcome'] == 'Unsuccessful':
        plt.plot((row['x'], row['endX']), (row['y'], row['endY']), color='red')
        plt.scatter(row['x'], row['y'], color='red')

# Set the x and y-axis limits
plt.xlim(0, 120)
plt.ylim(0, 80)

# Set the plot title
plt.title('Messi Pass Map vs Real Betis', color='white', size=20)

# Set equal aspect ratio for the plot
ax.set_aspect('equal')

# Show the plot
plt.show()
```
